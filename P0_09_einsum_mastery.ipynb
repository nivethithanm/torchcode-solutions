# P0-09: Einsum Mastery

**Difficulty**: 🟡 Medium

**Objective**: Translate ANY equation from a research paper directly to code using `torch.einsum`.

## The Rule
```
torch.einsum('subscripts of inputs -> subscripts of output', *tensors)
```
- Repeated letters = those dimensions are contracted (summed over)
- Letters only on input side = contracted; letters on both sides = kept

````python
import torch
````

## Task 1: Attention Score Computation

Given Q:(B,H,N,D) and K:(B,H,N,D), compute scores:(B,H,N,N)

````python
B, H, N, D = 2, 4, 8, 32
Q = torch.randn(B, H, N, D)
K = torch.randn(B, H, N, D)

# TODO: Write the einsum expression
scores = torch.einsum('bhnd,bhmd->bhnm', Q, K)  # torch.einsum('????', Q, K)

ref = Q @ K.transpose(-2, -1)
assert torch.allclose(scores, ref, atol=1e-5)
print('✅ Task 1 passed!')
````

## Task 2: Weighted Sum (Attention Output)

Given attn_weights:(B,H,N,N) and V:(B,H,N,D), compute output:(B,H,N,D)

````python
weights = torch.softmax(torch.randn(B, H, N, N), dim=-1)
V = torch.randn(B, H, N, D)

# TODO: Compute attention output
output = torch.einsum('bhnm,bhmd->bhnd', weights, V)  # torch.einsum('????', weights, V)

ref = weights @ V
assert torch.allclose(output, ref, atol=1e-5)
print('✅ Task 2 passed!')
````

## Task 3: Multi-Head Split Without View/Transpose

Project x:(B,N,D) with W:(H,D,Dk) to get heads:(B,H,N,Dk) using einsum.

````python
B, N, D, H, Dk = 2, 10, 64, 8, 8
x = torch.randn(B, N, D)
W = torch.randn(H, D, Dk)

# TODO: Project into heads using a single einsum
heads = torch.einsum('bnd,hdk->bhnk', x, W)  # torch.einsum('????', x, W)

assert heads.shape == (B, H, N, Dk)
print('✅ Task 3 passed!')
````

## Task 4: Bilinear Attention

Compute `score_ij = q_i^T W k_j` where q:(B,N,D), k:(B,M,D), W:(D,D)

````python
B, N, M, D = 2, 5, 7, 32
q = torch.randn(B, N, D)
k = torch.randn(B, M, D)
W = torch.randn(D, D)

# TODO: Compute bilinear scores → (B, N, M)
scores = torch.einsum('bnd,de,bme->bnm', q, W, k)  # torch.einsum('????', q, W, k)

# Verify against manual computation
ref = (q @ W) @ k.transpose(-2, -1)
assert torch.allclose(scores, ref, atol=1e-4)
print('✅ Task 4 passed!')
````
