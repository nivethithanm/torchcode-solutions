# B-01: Sigmoid & SiLU/Swish

**Difficulty**: 🟢 Easy | **Bridges**: Phase 1 activations → SwiGLU MLP

SiLU (Sigmoid Linear Unit) = Swish with β=1 = `x * sigmoid(x)`. This is THE activation in LLaMA, Mistral, DeepSeek.

```python
import torch
import torch.nn.functional as F
```

## Task 1: Implement Sigmoid

```python
def my_sigmoid(x):
    """σ(x) = 1 / (1 + exp(-x))"""
    # TODO: Implement
    return 1 / (1 + torch.exp(-x))

x = torch.randn(3, 4)
assert torch.allclose(my_sigmoid(x), torch.sigmoid(x), atol=1e-6)
print('✅ Sigmoid passed!')
```

## Task 2: Implement SiLU/Swish

```python
def my_silu(x):
    """SiLU(x) = x * sigmoid(x)"""
    # TODO: Implement using your sigmoid
    return x / (1 + torch.exp(-x))

x = torch.randn(3, 4)
assert torch.allclose(my_silu(x), F.silu(x), atol=1e-6)
print('✅ SiLU passed!')
```

## Task 3: Gating Mechanism

SwiGLU uses SiLU as a gate: `output = silu(x @ W_gate) * (x @ W_up)`. Implement this pattern.

```python
D, D_ff = 64, 256
x = torch.randn(2, 10, D)
W_gate = torch.randn(D_ff, D)
W_up = torch.randn(D_ff, D)

# TODO: Compute gated output
gate = my_silu(x @ W_gate.transpose(1, 0))   # silu(x @ W_gate.T)
up = x @ W_up.transpose(1, 0)     # x @ W_up.T
output = gate * up # gate * up

assert output.shape == (2, 10, D_ff)
print('✅ Task 3 passed! This is the core of SwiGLU.')
```
