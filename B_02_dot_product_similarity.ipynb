# B-02: Dot-Product Similarity

**Difficulty**: 🟢 Easy | **Bridges**: Linear algebra → Attention

Attention is just a weighted average where weights come from dot-product similarity.

```python
import torch
```

## Task 1: Pairwise Dot Products

```python
N, D = 5, 32
X = torch.randn(N, D)

# TODO: Compute NxN similarity matrix where sim[i,j] = X[i] · X[j]
similarity = X @ X.transpose(1, 0)  # YOUR CODE — one matrix operation

assert similarity.shape == (N, N)
# Diagonal should be the squared norms
for i in range(N):
    assert torch.allclose(similarity[i, i], (X[i] ** 2).sum(), atol=1e-5)
print('✅ Task 1 passed!')
```

## Task 2: Scaled Dot Product

Divide by sqrt(D) to prevent scores from growing too large.

```python
import math

Q = torch.randn(2, 8, 64)  # (B, N, D)
K = torch.randn(2, 8, 64)

# TODO: Compute scaled similarity: Q @ K^T / sqrt(D)
scores = Q @ K.transpose(-1, -2) / (Q.shape[-1] ** 0.5)  # YOUR CODE

assert scores.shape == (2, 8, 8)
# Scores should have roughly unit variance
print(f'Score variance: {scores.var():.2f} (should be ~1.0)')
print('✅ Task 2 passed!')
```
