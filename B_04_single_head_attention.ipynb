# B-04: Single-Head Attention

**Difficulty**: 🟡 Medium | **Bridges**: Basic ops → Scaled Dot-Product Attention

Implement the full attention mechanism for a SINGLE head before worrying about multi-head.

```python
import torch
import torch.nn as nn
import math
```

## Task: Complete Single-Head Attention

```python
class SingleHeadAttention(nn.Module):
    def __init__(self, d_model: int):
        super().__init__()
        # TODO: Create Q, K, V projection matrices (each d_model → d_model)
        self.Q = nn.Parameter(torch.randn(d_model, d_model))
        self.K = nn.Parameter(torch.randn(d_model, d_model))
        self.V = nn.Parameter(torch.randn(d_model, d_model))
        self.d_model = d_model

    def forward(self, x, mask=None):
        """Single-head self-attention.
        Args:
            x: (B, N, D)
            mask: optional (N, N) boolean mask
        Returns: (B, N, D)
        """
        B, N, D = x.shape
        assert D == self.d_model

        # TODO Step 1: Project to Q, K, V
        # TODO Step 2: Compute attention scores: Q @ K^T / sqrt(D)
        # TODO Step 3: Apply mask (if provided): masked_fill
        # TODO Step 4: Softmax
        # TODO Step 5: Weighted sum: attn_weights @ V

        q = x @ self.Q
        k = x @ self.K
        v = x @ self.V
        scores = q @ k.transpose(-2, -1) / math.sqrt(D)
        if mask is not None:
            scores = scores.masked_fill(mask == 0, float('-inf'))
        attn_weights = torch.softmax(scores, dim=-1)
        return attn_weights @ v

# Test
attn = SingleHeadAttention(64)
x = torch.randn(2, 10, 64)

# Without mask
out = attn(x)
assert out.shape == (2, 10, 64)

# With causal mask
mask = torch.tril(torch.ones(10, 10, dtype=torch.bool))
out_causal = attn(x, mask=mask)
assert out_causal.shape == (2, 10, 64)

print('✅ Single-Head Attention passed!')
```
